### Função de ativação sigmoide

$$f(x) = \frac{1}{1 + e^{-\beta \cdot x}}$$

- Implemente uma função em Python que retorne o valor da função sigmóide para uma dada entrada
- Nesta função, considere beta como parâmetro nomeado com valor igual a 1.0
- Plote o gráfico da função no intervalo de entrada previamente especificado
- Implemente uma função que retorne o valor da derivada da função sigmoide
- Em um mesmo gráfico, ilustre 3 valores distintos para $\beta$:
  - $\beta = 0.7$ na cor laranja
  - $\beta = 0.5$ na cor verde
  - $\beta = 0.3$ na cor vermelha
  - Use legendas no gráfico

```python
# Ativação sigmoide
def sigmoid(x, beta=1.0):
    return 1 / (1 + np.exp(-beta * x))
```

```python
# Derivada da função sigmoide
def sigmoid_derivative(x, beta=1.0):
    sig = sigmoid(x, beta)
    return beta * sig * (1 - sig)
```

```python
x = np.arange(-4, 4.1, 0.1)

# Valores de beta como solicitados
betas = [0.7, 0.5, 0.3]
colors = ['orange', 'green', 'red']
labels = [f'β={beta}' for beta in betas]

# Plot da função sigmoide
plt.figure(figsize=(10, 6))

for beta, color, label in zip(betas, colors, labels):
    y = sigmoid(x, beta)
    plt.plot(x, y, label=label, color=color, linestyle='-', linewidth=2)

plt.title('Função de Ativação Sigmoide - β')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend(loc='upper left')
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()

# Plot da derivada da função sigmoide
plt.figure(figsize=(10, 6))

for beta, color, label in zip(betas, colors, labels):
    dy = sigmoid_derivative(x, beta)
    plt.plot(x, dy, label=label, color=color, linestyle='--', linewidth=2)

plt.title('Derivada da Função Sigmoide - β')
plt.xlabel('x')
plt.ylabel('f\'(x)')
plt.legend(loc='upper left')
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```

### Função de Ativação ReLU (Unidade Linear Retificada)

$$f(x) = max(0,x)$$

- Implemente uma função em Python que retorne o valor da função ReLU para uma dada entrada
- Plote o gráfico da função no intervalo de entrada previamente especificado
- Implemente uma função que retorne o valor da derivada da função ReLU

```python
# Ativação ReLU
def relu(x):
    return np.maximum(0, x)
```

```python
# Derivada da ReLU
def relu_derivative(x):
    return np.where(x > 0, 1, 0)
```

```python
x = np.arange(-4, 4.1, 0.1)
relu_values = relu(x)
y_relu_derivative = relu_derivative(x)

# Plot da função ReLU
plt.figure(figsize=(10, 6))
plt.plot(x, relu_values, label='ReLU', color='blue')
plt.title('Função de Ativação ReLU')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.grid(True)
plt.show()

# Plot da derivada da ReLU
plt.figure(figsize=(10, 6))
plt.plot(x, y_relu_derivative, label='ReLU Derivada', color='red')
plt.title('Derivada da Função ReLU')
plt.xlabel('x')
plt.ylabel('f\'(x)')
plt.legend()
plt.grid(True)
plt.show()
```

### Função de ativação Tangente Hiperbólica

$$f(x) = tanh(x)$$

- Implemente uma função em Python que retorne o valor da função tangente hiperbólica para uma dada entrada
- Plote o gráfico da função no intervalo de entrada previamente especificado
- Implemente uma função que retorne o valor da derivada da função em questão
- Utilize bibliotecas para auxiliar nesta tarefa
- Considere o intervalo (-1,1)

```python
# Ativação Tangente Hiperbólica (tanh)
def tanh(x):
    return np.tanh(x)
```

```python
# Derivada da função tanh
def tanh_derivative(x):
    return 1 - np.tanh(x) ** 2
```

```python
# Intervalo (-1, 1)
x = np.arange(-1, 1.01, 0.01)
y_tanh = tanh(x)
y_tanh_derivative = tanh_derivative(x)

# Plot da função tanh
plt.figure(figsize=(10, 6))
plt.plot(x, y_tanh, label='Tangente Hiperbólica (tanh)', color='purple', linestyle='-', linewidth=2)
plt.title('Função de Ativação Tangente Hiperbólica (tanh)')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()

# Plot da derivada da função tanh
plt.figure(figsize=(10, 6))
plt.plot(x, y_tanh_derivative, label='Derivada da Tangente Hiperbólica (tanh)', color='orange', linestyle='-', linewidth=2)
plt.title('Derivada da Função Tangente Hiperbólica (tanh)')
plt.xlabel('x')
plt.ylabel('f\'(x)')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```

### Função de Ativação Gaussiana

$$f(x) = \exp \left( \frac{-(x - \mu)^2}{2\sigma^2} \right)$$

- $\mu$ é o ponto médio (centro) da curva
- $\sigma$ denota o raio de abertura
- Implemente uma função em Python que retorne o valor da função gaussiana para uma dada entrada
- Plote o gráfico da função no intervalo de entrada previamente especificado
- Implemente uma função que retorne o valor da derivada da função em questão


Considere também os seguintes valores ao plotar a função

- Parâmetros nomeados $\mu = 0$ e $\sigma = 1$
- $\mu = 0, \sigma = 0.3$
- $\mu = 0, \sigma = 0.5$
- $\mu = 0, \sigma = 0.7$
- $\mu = 0, \sigma = 0.9$
- $\mu = 0, \sigma = 1.1$

```python
# Ativação Gaussiana
def gaussian_func(x, mu_val=0, sigma_val=1):
    return np.exp(-((x - mu_val) ** 2) / (2 * sigma_val ** 2))
```

```python
# Derivada da função gaussiana
def gaussian_derivate(x, mu_val=0, sigma_val=1):
    return -(x - mu_val) / (sigma_val ** 2) * gaussian_func(x, mu_val, sigma_val)
```

```python
x = np.arange(-4, 4.1, 0.1)

# Parâmetros nomeados
mu_default = 0
sigma_default = 1

# mu e sigma
mu_values = [0, 0, 0, 0, 0, 0]
sigma_values = [1, 0.3, 0.5, 0.7, 0.9, 1.1]
colors = ['blue', 'orange', 'green', 'red', 'purple', 'cyan']
labels = [f'μ={mu}, σ={sigma}' for mu, sigma in zip(mu_values, sigma_values)]

# Plot da função Gaussiana
plt.figure(figsize=(10, 6))

for mu, sigma, color, label in zip(mu_values, sigma_values, colors, labels):
    y_gaussian = gaussian_func(x, mu, sigma)
    plt.plot(x, y_gaussian, label=label, color=color, linestyle='-', linewidth=2)

plt.title('Função de Ativação Gaussiana')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend(loc='upper right')
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()

# Plot da derivada da função Gaussiana
plt.figure(figsize=(10, 6))

for mu, sigma, color, label in zip(mu_values, sigma_values, colors, labels):
    y_gaussian_derivative = gaussian_derivate(x, mu, sigma)
    plt.plot(x, y_gaussian_derivative, label=label, color=color, linestyle='-', linewidth=2)

plt.title('Derivada da Função Gaussiana')
plt.xlabel('x')
plt.ylabel('f\'(x)')
plt.legend(loc='upper right')
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```

### Função de Ativação Leaky ReLU

$$f(x) = \left\{\begin{array}{cc} 0.01\cdot x & x < 0\\
x & x \geq 0
\end{array}
\right.$$

- Implemente uma função em Python que retorne o valor da função Leaky ReLU para uma dada entrada
- Plote o gráfico da função no intervalo de entrada previamente especificado
- Implemente uma função que retorne o valor da derivada da função Leaky ReLU (consulte a literatura!)

```python
# Ativação Leaky ReLU
def leaky_relu(x):
    return np.where(x >= 0, x, 0.01 * x)
```

```python
# Derivada da função Leaky ReLU
def leaky_relu_derivative(x):
    return np.where(x >= 0, 1, 0.01)
```

```python
input_range = np.arange(-4, 4.1, 0.1)
leaky_relu_values = leaky_relu(input_range)
leaky_relu_derivative_values = leaky_relu_derivative(input_range)

# Plot da função Leaky ReLU
plt.figure(figsize=(10, 6))
plt.plot(input_range, leaky_relu_values, label='Leaky ReLU', color='blue')
plt.title('Função de Ativação Leaky ReLU')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()

# Plot da derivada da função Leaky ReLU
plt.figure(figsize=(10, 6))
plt.plot(input_range, leaky_relu_derivative_values, label='Derivada da Leaky ReLU', color='red')
plt.title('Derivada da Função Leaky ReLU')
plt.xlabel('x')
plt.ylabel('f\'(x)')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```

```python
# Funções ReLU e Leaky ReLU
plt.figure(figsize=(10, 6))
plt.plot(x, relu_values, label='ReLU', color='blue', linestyle='-', linewidth=2)
plt.plot(x, leaky_relu_values, label='Leaky ReLU', color='red', linestyle='--', linewidth=2)
plt.title('Função de Ativação ReLU vs Leaky ReLU')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```
